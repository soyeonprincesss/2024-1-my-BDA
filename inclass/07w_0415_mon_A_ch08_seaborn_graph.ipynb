###시험예상문제

```python
import pandas as pd

# 중간고사 데이터 만들기
test1 = pd.DataFrame({'id'      : [1, 2, 3, 4, 5],
                      'midterm' : [60, 80, 70, 90, 85]})

# 기말고사 데이터 만들기
test2 = pd.DataFrame({'id'    : [1, 2, 3, 6, 7],
                      'final' : [70, 83, 65, 95, 80]})
```

```python
test1
```

```python
test2
```

```python
test1.to_html()
```

```python
from IPython.display import display_html
def display_side_by_side(*args):
    """여러 데이터프레임 비교가 쉽게 옆쪽으로 표시한다"""
    html_str=''
    for df in args:
        html_str += df.to_html() + '&nbsp;'*4
    display_html(html_str.replace('table','table style="display:inline"'), raw=True)
```

```python
display_side_by_side(test1, test2)
```

```python
display_side_by_side(test1, test2, test1.merge(test2)) #Inner join이랑 같음, 공통인것만 놓는것 Inner merge()임. #시험문제
```

```python
len(test1.merge(test2, how='inner')) #행의 수 ***1번문제
```

```python
test1.merge(test2, how='inner').iloc[1] #***2번문제
```

```python
test1.merge(test2, how='inner')
```

```python
display_side_by_side(test1, test2, test1.merge(test2, how='left')) #왼쪽에 있는거 그대로
```

```python
display_side_by_side(test1, test2, test1.merge(test2, how='right')) #오른쪽에있는거 그대로
```

```python
display_side_by_side(test1, test2, test1.merge(test2, how='outer')) #다 합쳐버리는것
```

```python
display_side_by_side(test1, test2, test1.merge(test2, how='cross')) #cross는 왼쪽에있는걸 오른쪽에있는 4개를 전부다 하나씩 다비교하는거임
```
