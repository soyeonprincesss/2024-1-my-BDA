## 데이터분석을 위한 seaborn 내장(토이) 데이터, titanic

```python
import seaborn as sns

titanic = sns.load_dataset("titanic")
titanic.head()
```

Titanic 데이터프레임에 대한 요약 정보를 제공

```python
titanic.info()
```

### 타이타닉 열(변수) 설명
- survived : 0 = 사망, 1 = 생존
- pclass : 1 = 1등석, 2 = 2등석, 3 = 3등석
- sex : male = 남성, female = 여성
- age : 나이
- sibsp : 타이타닉 호에 동승한 자매 / 배우자의 수
- parch : 타이타닉 호에 동승한 부모 / 자식의 수
- fare : 티켓 요금
- embarked : 탑승지, C = 셰르부르, Q = 퀸즈타운, S = 사우샘프턴
- class : First = 1등석, Second = 2등석, Third = 3등석
- who : 남/여/아이, 'man', 'woman', 'child'
- adult_male : 남자 어른, True/False
- deck : 방 위치, 'A', 'B', 'C', 'D', 'E', 'F', 'G', NaN
- embark_town : 탑승지
- alive : 생존, yes/no
- alone : 1인 탑승, True/False

##1. 성별 (sex)

```python
# 성별에 따른 데이터의 분포 시각화
sns.countplot(titanic, x = "sex");
```

```python
# 성별에 따른 생존률 시각화
sns.barplot(x="sex", y="survived", data=titanic); # 높이: 평균
```

```python
# Seaborn 라이브러리를 사용하여 타이타닉 데이터셋에서 성별에 따른 생존율의 평균을 시각화
sns.barplot(titanic, x="sex", y="survived", hue="sex");
```

```python
#각 성별 내에서 생존자와 사망자의 비율 시각화
sns.countplot(titanic, x = "who", hue = "alive", stat = "percent");
```

```python
#Titanic 데이터프레임을 'sex' 열을 기준으로 그룹화하고, 각 그룹에 대해 각 열의 값의 개수를 계산
titanic.groupby('sex').count()
```

```python
# 생존자와 사망자의 나이 분포를 박스플롯으로 비교
sns.boxplot(x='survived', y='age', data=titanic)
```

##3. 객실 등급 (class)

#분석 결과
성별에 따른 생존율: 여성의 생존율이 남성보다 높음.

나이에 따른 생존율: 생존자 그룹은 주로 20대 후반부터 30대 초반에 집중되어 있으며, 사망자 그룹은 20대 후반부터 40대까지 넓게 분포되어 있음.

객실 등급에 따른 생존율: 1등급 승객의 생존율이 가장 높고, 3등급 승객의 생존율이 가장 낮음.

나이에 따른 생존율: 어린이와 청년층의 생존율이 높음.

승선 항구에 따른 생존율: 특정 항구에서 승선한 승객의 생존율이 다름.

함께 탑승한 가족 수에 따른 생존율: 혼자 탑승한 경우보다 가족과 함께 탑승한 경우 생존 확률이 높음.

탑승 금액에 따른 생존율: 높은 탑승 금액을 지불한 승객들이 생존율이 높음.

탑승 항구에 따른 생존율: 각 항구의 지리적 위치나 승객들의 배정된 객실 위치와 관련된 요인들이 생존 여부에 영향을 줄 수 있음.

방위치에 따른 생존율: 객실이 위치한 데크가 생존 여부에 영향을 줄 수 있으며, 해당 정보는 생존 여부와의 상관 관계를 분석하는 데 도움이 될 수 있음.






#데이터 분석 결론
성별, 객실 등급, 나이, 승선 항구 등은 타이타닉 생존율에 중요한 영향을 미친다. 이러한 요인들을 통해 우리는 비슷한 상황에서의 대비책을 찾을 수 있다. 이러한 요인들을 종합적으로 고려하여 비상 시에는 여성과 어린이를 우선적으로 대피시키고, 상위 등급 객실을 우선적으로 이용하며, 어린이와 젊은 성인들을 우선적으로 구조하며, 승선 항구에 따라 대피 계획을 세우는 것 등이 중요하다.
